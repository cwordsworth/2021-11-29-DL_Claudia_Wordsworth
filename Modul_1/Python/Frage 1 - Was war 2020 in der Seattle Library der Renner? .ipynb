## Frage 1 - Wieviele Ausleihen gab es 2020 insgesamt, was waren die Top 5 Ausleihen und welche Materialientypen sind am beliebtesten?

```python
import urllib.request
import pandas as pd
```

```python
# Importieren der Daten und Dataframe erstellen 
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2020&$limit=30000"
checkout_file_2020 = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file_2020)
```

```python
checkouts_2020_df = pd.read_csv(checkout_file_2020)
```

```python
checkouts_2020_df
```

### Wieviele Ausleihen gab es insgesamt?

```python

checkouts_2020_df["checkouts"].sum()
```

### Was sind die meisten Ausleihen?

```python

checkouts_2020_df["checkouts"].max()
```

### Welche Materialtypen gibt es und wie oft kommen die einzelnen Materialtypen im Dataframe vor?

```python

checkouts_2020_df["materialtype"].unique() 
```

```python
checkouts_2020_df["materialtype"].value_counts()
```

### Was wurde am meisten ausgeliehen?

```python

checkouts_2020_df.sort_values("checkouts", ascending=False)
```

```python
# Am meisten wurden Kopfhörer ausgeliehen. Da das nicht so spannend ist, nehme ich den Materialtyp "Mixed" aus der Liste raus:
checkouts_materialtype_2020_df = checkouts_2020_df[checkouts_2020_df["materialtype"] != "MIXED"].sort_values("checkouts", ascending = False)
```

```python
# Die 5 am meisten ausgeliehen Ressourcen:
checkouts_materialtype_2020_df.head(5)
```

```python
checkouts_grouped_df = checkouts_materialtype_2020_df.groupby("materialtype").sum().sort_values("checkouts", ascending=False)
```
