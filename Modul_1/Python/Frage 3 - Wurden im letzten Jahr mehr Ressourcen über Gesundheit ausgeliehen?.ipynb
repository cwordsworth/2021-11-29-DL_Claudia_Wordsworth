##   Wurden im vergangenem Jahr mehr Ressourcen über Gesundheit ausgeliehen als vor 5 Jahren?

```python
import pandas as pd
import urllib.request
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2016&$limit=10000"
checkout_file_2016 = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file_2016)
```

```python
checkouts_2016_df = pd.read_csv(checkout_file_2016)
```

```python
checkouts_2016_df
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2020&$limit=10000"
checkout_file_2020 = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file_2020)
```

```python
checkouts_2020_df = pd.read_csv(checkout_file_2020)
```

```python
checkouts_2020_df
```

## Beide DF zu einer verbinden

```python
checkouts_2016_2020_df = pd.concat([checkouts_2020_df, checkouts_2016_df], ignore_index=True)
```

```python
checkouts_2016_2020_df
```

```python
checkouts_2016_2020_df.sort_values("checkouts", ascending=False).head(20)
```

### Wurden 2020 mehr Ressourcen über Gesundheit ausgeliehen?

```python
checkouts_health_df = checkouts_2016_2020_df[checkouts_2016_2020_df["title"].str.contains("Health")]
```

```python
checkouts_health_df.sort_values("checkouts", ascending=False)
```

```python
checkouts_health_df.groupby("checkoutyear").sum()[["checkouts"]]
```

```python
checkouts_2016_2020_df.dropna()
```

```python
checkouts_2016_2020_df.dropna(subset=["subjects"])
```

```python
checkouts_Health_subject_df = checkouts_2016_2020_df[checkouts_2016_2020_df['subjects'].str.contains("Health", case=False, na=False)]
```

```python
checkouts_Health_subject_df.sort_values("checkouts", ascending=False)
```

```python
checkouts_Health_subject_df.groupby("checkoutyear").sum()[["checkouts"]]
```

```python
%matplotlib inline
```

```python
checkouts_Health_subject_df.groupby("checkoutyear").sum()[["checkouts"]].plot(kind="bar")
```
