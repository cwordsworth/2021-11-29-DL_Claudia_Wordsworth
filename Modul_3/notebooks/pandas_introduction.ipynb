# Beispiele zur Auswahl und Erstellung von Spalten

```python
import pandas as pd
df = pd.read_csv("../data/Library_Usage.csv")
df.columns
```

```python
df
```

```python
#Beispiele wie man einzelne Spalten auswählt:
x = df['Total Renewals']
df[['Total Renewals', 'Total Checkouts']]
column_names = ['Total Renewals', 'Total Checkouts'] # auxiliary variable
subset = df[column_names]
print(subset)
print(x)
```

```python
#Spalten können mit einer Zuweisung (=) überschrieben oder neu erstellt werden:
df['dummy_variable'] = 5
df.head()
```

```python
x = df['dummy_variable']
df['dummy_variable'] = 10
x
```

```python
#Berechnungen auf schon bestehenden Variablen können auch direkt einer neuen Spalte zugeordnet werden:
import numpy as np

df['is_adult'] = df['Patron Type Definition'] == 'ADULT'
df['log_renewals'] = np.log(df['Total Renewals'] + 1)
```

```python
df.head()
```

```python
##2.5 Fallstudie: Feature Engineering:
#Die Spalte Circulation Active Year ist als Text und nicht als Zahl abgespeichert! Darum konvertiere Spalte in ein numerisches Format. 
#Überschreibe die ursprüngliche Variable mit den neuen Werten:
df["Circulation Active Year"]= pd.to_numeric(
  df['Circulation Active Year'], errors='coerce'
)
```

```python
df['Circulation Active Year'].dtypes
```

```python
#Ebenso müssen die Werte der Spalte Circulation Active Month in ein numerisches Datenformat umgewandelt werden: 
df['Circulation Active Month'] = pd.to_datetime(
    df['Circulation Active Month'],
    errors='coerce',
    format="%B"
)

```

```python
#Jetzt extrahieren wir den Monat als Zahl aus der Spalte
df['Circulation Active Month']= df['Circulation Active Month'].dt.month
df['Circulation Active Month']
```

```python
#Berechnen der aktiven Mitgliedschaftsdauer in Monaten speichere das Ergebnis in neuer Spalte Membership Duration:
df["Membership Duration"]=(df['Circulation Active Year'] - df['Year Patron Registered'])*12 + df['Circulation Active Month']
```

```python
#Fehlende Werte in 0 ändern:
df["Membership Duration"]=df["Membership Duration"].fillna(0)
```

```python
df.head(10)
```

```python
#Auswahl von Zeilen über logische Ausdrücke mittels loc[]:
df.loc[df['Total Checkouts'] > 10000].head()
```

```python
#Mithilfe einer Hilfsvariable, die den booleschen Vektor zwischenspeichert:
row_filter = df['Total Checkouts'] > 10000
df.loc[row_filter].head()
```

```python
#Mit booleschen Zeilenfiltern können komplexe logische Ausdrücke unter Zuhilfenahme von Operatoren dargestellt werden:
row_filter = (df['Patron Type Definition'] == 'SENIOR') & (df['Notice Preference Definition'] == 'email')
df.loc[row_filter]
row_filter
```

```python
#Funktion Series.between(left, right), mit der eine boolesche Series erstellt wird, die True ist 
#wenn der Wert der ursprünglichen Series zwischen oder auf den Werten left und right liegt:
filter1 = (df['Total Checkouts'] >= 20) & (df['Total Checkouts'] <= 80)
filter2 =  df['Total Checkouts'].between(20, 80)
all(filter1 == filter2)
```

# Behandlung von Fehlenden Werten

```python
#Filtern und Umschreiben von fehlenden Werten:
df[df['Age Range'].isna()]
df[df['Age Range'].notna()]
```

```python
#Anzahl der Werte:
df.isna().sum()
```

```python
#Entfernen fehlender Werte mit dropna(). Originalvariable bleibt unverändert. Kann mit inplace=True verändert werden:
# drops all rows that contain at least one missing values
df.dropna()
# drops all missing values in this series
df['Age Range'].dropna()
```

```python
#Ersetzen mit fillna(). Originalvariable bleibt unverändert. Kann mit inplace=True verändert werden:
df['Age Range'].fillna("keine Angabe")
```
