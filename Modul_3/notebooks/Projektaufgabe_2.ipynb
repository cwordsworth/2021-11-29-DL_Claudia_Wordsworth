```python
import pandas as pd
Lib_Use = pd.read_csv("../data/Library_Usage.csv", na_values="None")

Lib_Use.head(10)
```

Der Datensatz ist eine Vollstichprobe mit 15 verschiedenen Merkmalen. Die Merkmale sind nominal (Home Library Definition), metrisch (Total Checkouts) und kategorisch (Age Range).

```python
Lib_Use.dropna(inplace=True)
```

```python
Lib_Use
```

```python
#Berechnung der neuen Variablen Membership Duration:

Lib_Use['Circulation Active Month'] = pd.to_datetime(
    Lib_Use['Circulation Active Month'],
    errors='coerce',
    format="%B"
)
```

```python
Lib_Use['Circulation Active Month']= Lib_Use['Circulation Active Month'].dt.month
Lib_Use["Membership Duration"]=(Lib_Use['Circulation Active Year'] - Lib_Use['Year Patron Registered'])*12 + Lib_Use['Circulation Active Month']
```
